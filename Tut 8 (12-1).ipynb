```python
import pandas as pd

checkouts_df = pd.read_csv('seattle_checkouts.csv')
checkouts_df
```

Make a pie chart showing the most common publication years in the list. 


```python
#What's the error here?
checkouts_df["PublicationYear"].valuecounts()
```

```python
plot = checkouts_df['PublicationYear'].value_counts().plot(kind="barh", figsize=(5, 5))
print(plot)
```

Add a column that checks if the given title is a mystery.

```python
checkouts_df["Mystery?"] = checkouts_df["Subjects"].str.contains("Mystery")
checkouts_df
```

Make a pie chart!

```python
plot = checkouts_df["Mystery?"].value_counts().plot(kind="pie", figsize=(5, 5), colors = ["rosybrown", "bisque", "#aaaaff"])
print(plot)
```

Unpack the stack of methods in the following lines.

```python
nytg_df = pd.read_csv('nyt_full_gender_signal.tsv', sep="\t")
nytg_df
```

```python
#remember: Python evaluates a stack of methods from left to right!

nytg_df.groupby(['year', 'gender_signal']).size().unstack(fill_value=0)
```
