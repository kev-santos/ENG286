Example: Let's try to analyze the lines of the poem "Song of Myself" by Walt Whitman.

```python
poem = open("song_of_myself.txt", encoding="utf-8").read()
```

```python
# Split up the poem by line. What should I put as an argument for the .split() method?
poem_lines = poem.split("\n")
poem_lines
```

```python
len(poem_lines)
# Does this accurately give me the number of lines in the poem?
```

```python
poem_lines.remove('')
poem_lines
```

```python
poem_lines_nonempty = []

for line in poem_lines:
    if line != '':
        poem_lines_nonempty.append(line)
        
poem_lines_nonempty        
```
