Add a column that checks if the given title is a mystery.

```python
checkouts_df["Mystery?"] = checkouts_df["Subjects"].str.contains("Mystery")
checkouts_df
```

Make a pie chart!

```python
plot = checkouts_df["Mystery?"].value_counts().plot(kind="pie", figsize=(10, 10))
print(plot)
```
